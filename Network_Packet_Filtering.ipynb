```python
# Drop IP addresses and other non-numeric fields that are not useful for modeling
df.drop(['Source IP', 'Destination IP', 'Timestamp'], axis=1, inplace=True)
```

```python
# Encode categorical variables like Protocol and Flags
df['Protocol'] = df['Protocol'].astype('category').cat.codes
df['Flags'] = df['Flags'].astype('category').cat.codes
```

```python
# Separate features (X) and labels (y)
X = df.drop('Intrusion', axis=1)
y = df['Intrusion']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Normalize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define the models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=1000),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100)
}

# Train, predict, and evaluate each model
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Accuracy for {name}: {accuracy_score(y_test, y_pred)}")
    print(f"Classification Report for {name}:\n{classification_report(y_test, y_pred)}\n")
    
    filename = f'{name.replace(" ", "_").replace("(", "").replace(")", "")}.pkl'
    joblib.dump(model, filename)
    print(f"Model saved as {filename}\n")
```
